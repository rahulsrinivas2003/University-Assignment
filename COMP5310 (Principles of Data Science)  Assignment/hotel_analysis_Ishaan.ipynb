## Hotel Booking

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hotel = pd.read_csv("hotel_bookings.csv")
hotel.head()
```

## Ananlysis of Data Structure

```python
hotel.info()
```

```python
hotel = hotel.rename(columns={"adr":"average_daily_room_rate"})
```

```python
hotel.describe()
```

```python
hotel.info()
```

```python
missing = hotel.isnull().sum()
missing_pct = (missing / len(hotel) * 100).round(2)
missing_hotel = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_hotel = missing_hotel[missing_hotel['Missing Count'] > 0].sort_values('Missing %', ascending=False)
print(missing_hotel.to_string())

# Visualise
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(missing_hotel.index, missing_hotel['Missing %'], color='blue')
ax.set_xlabel('Missing (%)')
ax.set_title('Missing Value Proportions by Feature')
for i, v in enumerate(missing_hotel['Missing %']):
    ax.text(v + 0.3, i, f'{v:.1f}%', va='center', fontsize=9)
plt.tight_layout()
plt.show()
```

## Data Cleaning 
Since most of the data is company is missing we can drop that column

```python
hotel.shape
```

```python
hotel.drop(columns=["company"], inplace=True)
```

```python
hotel = hotel.apply(lambda col: col.str.strip() if col.dtype == "object" or col.dtype.name == "string" else col)
```

```python
hotel.info()
```

```python
for col in hotel:
    print(col, hotel[col].unique())
```

```python
hotel["hotel"] = hotel["hotel"].replace("CityHotel", "City Hotel").replace("ResortHotel", "Resort Hotel")
```

```python
hotel["hotel"].unique()
```

```python
hotel["market_segment"] = hotel["market_segment"].str.title()
hotel["market_segment"] = hotel["market_segment"].replace("Direc", "Direct").replace("Offline Ta/To", "Offline TA/TO").replace("Online Ta", "Online TA").replace("Corporate ", "Corporate").replace("Undefined ", "Undefined")
hotel = hotel.dropna()
```

```python
hotel["market_segment"].unique()
```

```python
for col in hotel:
    print(col, hotel[col].unique())
```

```python
hotel["distribution_channel"] = hotel["distribution_channel"].replace("Direct ", "Direct").replace("Corporate ", "Corporate").replace("gds", "GDS")
hotel["distribution_channel"].unique()
```

```python
hotel["meal"] = hotel["meal"].replace("Undefined", "SC").replace("SC ", "SC").replace("bb", "BB").replace("FB ", "FB").replace("Undefined ", "SC").replace("HB ", "HB")
```

```python
hotel["meal"].unique()
```

```python
hotel["customer_type"] = hotel["customer_type"].str.title().replace("Transient ", "Transient").replace("Transient-Party ", "Transient-Party")
```

```python
hotel["customer_type"].unique()
```

```python
# handling missing values
hotel['agent'] = hotel['agent'].fillna(0).astype(int)
hotel['children'] = hotel['children'].fillna(0).astype(int)
```

```python
hotel['country'] = hotel['country'].fillna('Unknown')
```

```python
hotel = hotel.drop_duplicates()
```

```python
# removing data with no guests
hotel = hotel[(hotel["adults"] + hotel["children"] + hotel["babies"]) > 0]
```

```python
hotel.info()
```

```python
hotel["adults"].sum()
```

```python
hotel = hotel[hotel["average_daily_room_rate"]>0]
```

```python
hotel.shape
```

```python
hotel.head()
```

```python
hotel["reservation_status_date"] = pd.to_datetime(hotel["reservation_status_date"], format="mixed", dayfirst=True)
```

```python
hotel.head()
```

```python
hotel.info()
```

```python
month_map = {
    "January": 1, "February": 2, "March": 3,
    "April": 4, "May": 5, "June": 6,
    "July": 7, "August": 8, "September": 9,
    "October": 10, "November": 11, "December": 12
}
hotel["arrival_date_month"] = hotel["arrival_date_month"].map(month_map)
```

```python
hotel.head()
```

```python
hotel.info()
```

```python
hotel["reservation_status"] = hotel["reservation_status"].str.title().replace("No-Show ", "No-Show")
```

```python
hotel["reservation_status"].unique()
```

# New Features

```python
# total_nights column creation
hotel["total_nights"] = hotel["stays_in_weekend_nights"] + hotel["stays_in_week_nights"]
```

```python
# total_guests column creation
hotel["total_guests"] = hotel["adults"] + hotel["children"] + hotel["babies"]
```

```python
hotel.tail()
```

```python
# arival_date column creation
hotel["arrival_date"] = pd.to_datetime(hotel["arrival_date_year"].astype(str) + "-" +hotel["arrival_date_month"].astype(str) + "-" +hotel["arrival_date_day_of_month"].astype(str))
```

```python
hotel.head()
```

```python
hotel.describe()
```

```python
# Canceled Labels creation
canceled_labels = {0: 'Not Cancelled', 1: 'Cancelled'}
hotel['cancel_label'] = hotel['is_canceled'].map(canceled_labels)
```

## EDA for Lead-Time Numberic Feature

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 8))

# Histogram
axes[0][0].hist(hotel['lead_time'], bins=30)
axes[0][0].set_title('Distribution of Lead Time')
axes[0][0].set_xlabel('Lead Time (days)')
axes[0][0].set_ylabel('Count')

# Boxplot by cancellation
hotel.boxplot(column='lead_time', by='cancel_label', ax=axes[1][0])
axes[1][0].set_title('Lead Time by Cancellation Status')
axes[1][0].set_xlabel('Cancellation Status')
axes[1][0].set_ylabel('Lead Time (days)')

# Simple boxplot
axes[0][1].boxplot(hotel["lead_time"])
axes[0][1].set_title("Lead Time Boxplot")
axes[0][1].set_xlabel('Lead Time')
axes[0][1].set_ylabel('Days')

# Count plot (bar chart)
axes[1][1].bar(
    hotel["cancel_label"].value_counts().index,
    hotel["cancel_label"].value_counts().values
)
axes[1][1].set_title("Cancellation Count")
axes[1][1].set_xlabel('Cancellation Status')
axes[1][1].set_ylabel('Number of Bookings')

# Remove default pandas title
plt.suptitle('')
plt.tight_layout()
plt.show()
```

## EDA for ADR Numeric Feature

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Histogram of ADR (cliping extreme outliers for readability)
adr_clip = hotel['average_daily_room_rate'].clip(upper=500)
axes[0].hist(adr_clip, bins=60, color='coral', edgecolor='white')
axes[0].set_title('Distribution of ADR (clipped at 500)')
axes[0].set_xlabel('Average Daily Rate ($)')
axes[0].set_ylabel('Count')

# Scatter: lead_time vs adr, coloured by cancellation
colors = hotel['is_canceled'].map({0: 'steelblue', 1: 'coral'})
axes[1].scatter(hotel['lead_time'], hotel['average_daily_room_rate'].clip(upper=500), 
                c=colors, alpha=0.15, s=5)
axes[1].set_title('Lead Time vs ADR (blue=kept, orange=cancelled)')
axes[1].set_xlabel('Lead Time (days)')
axes[1].set_ylabel('ADR ($)')

plt.tight_layout()
plt.show()
```

## Heatmap for correlation

```python
corr = hotel[['lead_time', 'average_daily_room_rate', 'is_canceled', 'total_nights', 'total_of_special_requests']].corr()
print('Correlation matrix:')
print(corr.round(3))
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(7, 5))
num_cols = ['lead_time', 'average_daily_room_rate', 'is_canceled', 'total_nights',
            'total_of_special_requests', 'previous_cancellations', 'booking_changes']
sns.heatmap(hotel[num_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm',
            center=0, ax=ax, linewidths=0.5)
ax.set_title('Correlation Heatmap — Key Numerical Features')
plt.tight_layout()
plt.show()
```

## EDA for Categorical Feature Market_Segment

```python
seg_cancel = (hotel.groupby('market_segment')['is_canceled']
                .agg(['mean', 'count'])
                .rename(columns={'mean': 'cancellation_rate', 'count': 'n_bookings'})
                .sort_values('cancellation_rate', ascending=False)
                .reset_index())

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar: cancellation rate
axes[0].barh(seg_cancel['market_segment'], seg_cancel['cancellation_rate'] * 100,
             color='steelblue')
axes[0].set_xlabel('Cancellation Rate (%)')
axes[0].set_title('Cancellation Rate by Market Segment')
for i, v in enumerate(seg_cancel['cancellation_rate']):
    axes[0].text(v * 100 + 0.3, i, f'{v*100:.1f}%', va='center', fontsize=9)

# Bar: volume of bookings
axes[1].barh(seg_cancel['market_segment'], seg_cancel['n_bookings'],
             color='coral')
axes[1].set_xlabel('Number of Bookings')
axes[1].set_title('Booking Volume by Market Segment')

plt.tight_layout()
plt.show()
print(seg_cancel.to_string(index=False))
```

## EDA for Categorical Feature Deposit_Type

```python
dep_cancel = (hotel.groupby('deposit_type')['is_canceled']
                .agg(['mean', 'count'])
                .rename(columns={'mean': 'cancellation_rate', 'count': 'n_bookings'})
                .sort_values('cancellation_rate', ascending=False)
                .reset_index())

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

# Grouped bar: cancellation rate
bars = axes[0].bar(dep_cancel['deposit_type'], dep_cancel['cancellation_rate'] * 100,
                   color=['#e07b54', '#5b8db8', '#6aab6a'])
axes[0].set_ylabel('Cancellation Rate (%)')
axes[0].set_title('Cancellation Rate by Deposit Type')
for bar, v in zip(bars, dep_cancel['cancellation_rate']):
    axes[0].text(bar.get_x() + bar.get_width()/2, v*100 + 0.5,
                 f'{v*100:.1f}%', ha='center', fontsize=10)

# Stacked bar: composition (cancelled vs kept)
kept = 1 - dep_cancel['cancellation_rate']
axes[1].bar(dep_cancel['deposit_type'], dep_cancel['cancellation_rate'] * 100,
            label='Cancelled', color='coral')
axes[1].bar(dep_cancel['deposit_type'], kept * 100,
            bottom=dep_cancel['cancellation_rate'] * 100,
            label='Kept', color='steelblue')
axes[1].set_ylabel('%')
axes[1].set_title('Booking Outcome Composition by Deposit Type')
axes[1].legend()

plt.tight_layout()
plt.show()
print(dep_cancel.to_string(index=False))
```
